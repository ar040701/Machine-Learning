```python
#feature_selection using pearsons coefficient
plt.figure(figsize=(12,10))
cor=data.corr()
sns.heatmap(cor,annot=True,cmap=plt.cm.CMRmap_r)
```

```python
#with the following feature we can select highly correlated features
#it will remove the first feature that iscorrelated with any other feature
def correlation(data,thresold):
    col_corr=set()
    corr_matrix=data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>thresold:#getting absolute coefficient value
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr           
```

```python
corr_features=correlation(x,0.8)
len(set(corr_features))
```

```python
corr_features   #these are the features that are highly correlated with higher correlation value
```
